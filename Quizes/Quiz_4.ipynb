```python
root_dir = r'C:/Users/prady/Downloads/text_data/text/'
```

```python
files = os.listdir(r'C:/Users/prady/Downloads/text_data/text/')
print("Numbers of files :", len(files))
```

```python
df = pd.DataFrame(columns=['label','text'])
for file in files:
    with open(os.path.join(root_dir,file),'rt') as f:
        t = f.readlines()
        df = df.append({
            'label': t[0].strip('\n'),
            'text': t[1]
        }, ignore_index=True)
```
